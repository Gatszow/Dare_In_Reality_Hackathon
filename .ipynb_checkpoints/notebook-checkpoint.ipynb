## Converting to seconds

```python
def seconds_conversion(input_time):
  if pd.isna(input_time):
    return input_time
  else:
    if len(input_time.split(':'))==2 and '.' in input_time.split(':')[1]:
      millsecs_1 = int(input_time.split(':')[0])*60*1000
      millsecs_2 = int(input_time.split(':')[1].split('.')[0])*1000 + int(input_time.split(':')[1].split('.')[1])
      final_conversion = millsecs_1 + millsecs_2
    else:
      final_conversion = int(input_time.split(':')[0])*60*1000 + int(input_time.split(':')[1])*1000

  return final_conversion
```

```python
train_data['s1_large'] = train_data.copy()['s1_large'].apply(milliseconds_conversion)
train_data['s2_large'] = train_data.copy()['s2_large'].apply(milliseconds_conversion)
train_data['s3_large'] = train_data.copy()['s3_large'].apply(milliseconds_conversion)
train_data['elapsed'] = train_data.copy()['elapsed'].apply(milliseconds_conversion)
train_data['hour'] = train_data.copy()['hour'].apply(milliseconds_conversion)

test_data['s1_large'] = test_data.copy()['s1_large'].apply(milliseconds_conversion)
test_data['s2_large'] = test_data.copy()['s2_large'].apply(milliseconds_conversion)
test_data['s3_large'] = test_data.copy()['s3_large'].apply(milliseconds_conversion)
test_data['elapsed'] = test_data.copy()['elapsed'].apply(milliseconds_conversion)
test_data['hour'] = test_data.copy()['hour'].apply(milliseconds_conversion)
```

```python
def replace_first_comma_and_delete_others(s):
    count = s.count(',')
    for i in range(count-1):
        position = s.rfind(',')
        s = s[:position] + s[position+1:]
    return s.replace(',', '.')
```

```python
def replace_commas_in_columns_in_df(df, columns):
    replaced_df = df.copy()
    for column in columns:
        replaced_df[column] = replaced_df[column].apply(replace_first_comma_and_delete_others)
    return replaced_df
```

```python
list_of_columns = ['air_temp', 'track_temp', 'humidity', 'pressure', 'wind_speed']
data_weather = replace_commas_in_columns_in_df(data_weather, list_of_columns)
data_weather
```

# Creating Models:

```python
# metric
from sklearn.metrics import mean_squared_log_error

def check_metric_score(y_true, y_pred):
    return round(mean_squared_log_error(y_true, y_pred), 5)

def metric_score_for_grid(estimator, X, y):
    return check_metric_score(y, estimator.predict(X))
```

```python
def check_model_score(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    return check_metric_score(y_test, model.predict(X_test))

def check_model_score_tf(nn, X_train, y_train, X_test, y_test, loss=tf.keras.losses.MeanSquaredLogarithmicError(), optimizer='Adam', metrics=None, epochs=100, callbacks=None, verbose=0):
    nn.compile(loss=loss,
               optimizer=optimizer,
               metrics=metrics)
    nn.fit(X_train, y_train, epochs=epochs, verbose=verbose, callbacks=callbacks)
    return check_metric_score(y_test, nn.predict(X_test))
```

```python
from sklearn.model_selection import KFold 
def check_kfold_score(model, X, y, k=3):
    kf = KFold(n_splits=k)
    model_scores = []
    for train_index, test_index in kf.split(min_max_train_df):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index] 
        y_train, y_test = y[train_index], y[test_index]

        model_scores.append(check_model_score(model, X_train, y_train, X_test, y_test))
        
    return model_scores

def check_kfold_score_tf(nn, X, y, loss=tf.keras.losses.MeanSquaredLogarithmicError(), optimizer='Adam', metrics=None, epochs=100, callbacks=None, k=3, verbose=0):
    kf = KFold(n_splits=k)
    model_scores = []
    
    fold_no = 1
    for train_index, test_index in kf.split(min_max_train_df):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index] 
        y_train, y_test = y[train_index], y[test_index]
        
        print('------------------------------------------------------------------------')
        print(f'Training for fold {fold_no} ...')
        model_scores.append(check_model_score_tf(nn, X_train, y_train, X_test, y_test, loss=loss, optimizer=optimizer, metrics=metrics, epochs=epochs, callbacks=callbacks, verbose=verbose))
        fold_no += 1
        
    return model_scores
```

```python
# hyperparameter tuning
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

def find_the_best_parameters(model, X, y, parameters, k=3):
    grid = GridSearchCV(model, parameters, cv=k, scoring=make_scorer(check_metric_score, greater_is_better=False))
    grid.fit(X, y)
    print(grid.best_estimator_, grid.best_score_)
```
